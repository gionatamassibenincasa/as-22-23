Queste sono utili durante lo sviluppo di exploit in quando ti permettono di convertire per esempio indirizzi di memoria in forma numerica nella loro rappresentazione in bytes in little o big endian.
<br><br>
Pwntools offre dei wrapper per la libreria <code class="code">struct</code> di python (che offre <code class="code">struct.pack</code>).</p>
<ul>
<li><code class="code">p64(num, endianness="little", ...)</code> Esegue il packing di un integer a 64 bit</li>
<li><code class="code">p32(num, endianness="little", ...)</code> Esegue il packing di un integer a 32 bit</li>
<li><code class="code">u64(data, endianness="little", ...)</code> Esegue l'unpacking di integer a 64 bit</li>
<li><code class="code">u32(data, endianness="little", ...)</code> Esegue l'unpacking di integer a 32 bit</li>
</ul>
<p>Per esempio:</p>
<ul>
<li><code class="code">p64(0x401020) -&gt; b"\x20\x10\x40\x00\x00\x00\x00\x00"</code></li>
<li><code class="code">u32(b"\x00\x50\x40\x00") -&gt; 0x405000</code></li>
</ul>
<p>Puoi trovare maggiorni informazioni nella <a href="https://docs.pwntools.com/en/stable/util/packing.html" target="_blank">documentazione relativa</a>
